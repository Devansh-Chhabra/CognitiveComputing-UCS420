Q1)

For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'). Find
1. Sum of all elements
2. Sum of all elements row-wise
3. Sum of all elements column-wise

```python
import numpy as np
gfg = np.matrix([[4,1,9],[12,3,1],[4,5,6]])
print(gfg,end='\n\n')

print("Sum of all elements:",np.sum(gfg),end='\n\n')

for i in range(gfg.shape[0]):
  ans = 0
  for j in range(0,gfg.shape[1]):
    ans += gfg[i,j]
  print("Sum of row",i+1,"is:",ans)

print()

for i in range(gfg.shape[1]):
  ans = 0
  for j in range(0,gfg.shape[0]):
    ans += gfg[j,i]
  print("Sum of col",i+1,"is:",ans)
```

Q3)

You are given a weekly sales dataset and need to perform various operations
using NumPy broadcasting.
1. Generate your unique sales dataset:
  * Take the sum of the ASCII values of the initials of your first and last name. Call this value X. (If your initials are A B → ASCII sum = 65+66 = 131 → sales = [131, 181, 231, 281, 331].)
  * Create a NumPy array sales with values [X, X+50, X+100, X+150,
X+200].

2. Compute your personalized tax rate as ((X % 5) + 5) / 100.
  * Use broadcasting to apply this tax rate to each sales value.

3. Adjust sales based on discount:
  * If sales < X+100, apply a 5% discount.
  * If sales >= X+100, apply a 10% discount.

4. Expand sales data for multiple weeks:
  * Create a 3×5 matrix representing three weeks of sales by stacking
sales three times using broadcasting.
  * Increase sales by 2% per week using element-wise broadcasting.

```python
# Devansh Chhabra
X = ord('D') + ord('C')
print(X,end='\n\n')

sales = np.array([X,X+50,X+100, X+150,X+200])
print(sales,end='\n\n')

tax = ((X % 5) + 5) / 100
sales = tax + sales
print(sales,end='\n\n')

index = 0
while(index < len(sales)):
  if(sales[index] < X+100):
    sales[index] = 0.95 * sales[index]
  else:
    sales[index] = 0.9 * sales[index]
  index += 1

expandedSales = np.tile(sales,(3,1))
print(expandedSales,end='\n\n')

weeks = np.arange(3).reshape(3,1)
multiplier = 1 + weeks * 0.02
expandedSales *= multiplier
print(expandedSales,end='\n\n')
```

Q4)

Generate x values using np.linspace() from -10 to 10 with 100 points. Use
each function from the list below and compute y values using NumPy:
  * Y = x2
  * Y = sin(x)
  * Y = ex
  * Y = log(|x| + 1)
  
Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.

```python
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

plt.plot(x, y1, label='Y = x^2')
plt.plot(x, y2, label='Y = sin(x)')
plt.plot(x, y3, label='Y = e^x')
plt.plot(x, y4, label='Y = log(|x| + 1)')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Plots of Different Mathematical Functions")
plt.legend()
plt.grid(True)
plt.show()
```
