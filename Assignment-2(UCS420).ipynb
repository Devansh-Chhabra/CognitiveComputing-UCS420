**QUESTION-3**

WAP to create a list of 100 random numbers between 100 and 900. Count and print the:
1. All odd numbers
2. All even numbers
3. All prime numbers

```python
# Ques-03
import random

List1 = [random.randint(100,900) for i in range(100)]
print(List1)

oddList = []
evenList = []
primeList = []

for i in List1:
  if(i % 2 == 0):
    evenList = evenList + [i]
  else:
    oddList = oddList + [i]

    flag = True;
    for j in range(3,i,2):
      if(i % j == 0):
        flag = False;
        break;

    if(flag == True):
      primeList = primeList + [i]

print("")
print("Total Odd Elements:",len(oddList),"->",oddList)
print("Total Even Elements:",len(evenList),"->",evenList)
print("Total Prime Elements:",len(primeList),"->",primeList)
```

**QUESTION-04**

Consider the following two sets, A and B, representing scores of two teams in multiple matches. A = {34, 56, 78, 90} and B = {78, 45, 90, 23}

WAP to perform the following operations using set functions:
  1. Find the unique scores achieved by both teams (union of sets).
  2. Identify the scores that are common to both teams (intersection of sets).
  3. Find the scores that are exclusive to each team (symmetric difference).
  4. Check if the scores of team A are a subset of team B, and if team B's scores are a superset of team A.
  5. Remove a specific score X (input by the user) from set A if it exists.  If not, print a message saying it is not present.

```python
# Ques-04

A = {34,56,78,90}
B = {78,45,90,23}

uniqueScores = A.union(B)
print(uniqueScores)

commonScores = A.intersection(B)
print(commonScores)

exclusiveScores = A.symmetric_difference(B)
print(exclusiveScores)

isSubset = A.issubset(B)
print(isSubset)

X = int(input("Enter the score: "))
if X in A:
  A.remove(X)
  print("Score Removed")
else:
  print("Score is not present")
```
