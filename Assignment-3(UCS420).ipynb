**QUESTION-6**

Create a sample dataset (employees.csv) containing information about employees in a company.

```python
from google.colab import files
uploaded = files.upload()

import numpy as np
import pandas as pd
df3 = pd.read_csv('employees.csv')
```

**QUESTIONS**

1. Shape (number of rows and columns) of the DataFrame.
2. Summary of the DataFrame that includes the data types and non-null counts for
each column.
3. Generate descriptive statistics.
4. Display the first 5 rows and last 3 rows of the dataset.

```python
# Ques-01
print("Shape of the dataFrame:",df3.shape,end='\n\n')

#Ques-02
print("Printing the info about the DataFrame")
df3.info()
print(end='\n\n')

# Ques-03
print("Printing the descriptive Stats")
print(df3.describe())
print(end='\n\n')

# Ques-04
print(df3.head(5),end='\n\n')
print(df3.tail(3),end='\n\n')
```

5. Calculate the following statistics from the dataset:
  1. The average salary of employees.
  2. The total bonus paid to all employees.
  3. The youngest employee's age.
  4. The highest performance rating.

6. Sort the DataFrame by the Salary column in descending order.

7. Add a new column that categorizes employees based on their performance rating:
  1. Excellent for ratings >= 4.5
  2. Good for ratings >= 4.0 but < 4.5
  3. Average for ratings < 4.0
  
8. Identify missing values in the DataFrame.

```python
# Ques-05
print("Average Salary:",df3["Salary"].mean())
print("Bonus paid:",df3["Bonus"].sum())
print("Youngest Employee's age:",df3["Age"].min())
print("Highest Performance Rating:",df3["Rating"].max(),end='\n\n\n')

# Ques-06
print(df3.sort_values(by="Salary",ascending = False),end='\n\n\n')

# Ques-07

# Manual Way
# df3["Performance Category"] = ["Excellent","Good","Average","Excellent","Average"]

# Function
def performance(rating):
  if rating >= 4.5:
    return "Excellent"
  elif rating >= 4.0:
    return "Good"
  else:
    return "Average"

df3["Performance Category"] = df3["Rating"].apply(performance)
print(df3,end='\n\n\n')

# Ques-08
print("Print Missing Values")
print(df3.isnull(),end='\n\n')
```

9. Rename the Employee_ID column to ID.

10. Find all employees who:
  1. Have more than 5 years of experience.
  2. Belong to the IT department.

11. Modify the dataset by adding a new column, Tax, which deducts 10% of the
  Salary.
  
12. Save the modified DataFrame (with added columns) to a new CSV file.

```python
# Ques-09
df3.rename(columns = {"Employee_ID": "ID"},inplace = True)

# Ques-10
print("Employees having experience of more than 5 years and belong to IT department",end='\n\n')
print(df3[(df3["Years_of_Experience"] > 5) & (df3["Department"] == "IT")],end='\n\n')

# Ques-11
df3["Tax"] = df3["Salary"] * 0.10
print("Printing the updated Data",end='\n\n')
print(df3)

# Ques-12
finaldoc = "updated_employees.csv"
df3.to_csv(finaldoc,index = False)
```
